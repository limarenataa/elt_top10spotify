**EXTRACT**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('spotify-2023.csv', sep=",", encoding='latin-1')


```

```python
df
```

```python
dadosSpotifySelecionado = df.filter(items=['track_name', 'artist(s)_name', 'released_year', 'streams', 'danceability_%', 'energy_%'])
```

**TRANSFORM**

```python
dadosSpotifySelecionado.columns=['Música', 'Artista','Lançamento', 'Reproduções', 'Dançabilidade', 'Energia']
```

```python
top10MusicasDancantes = dadosSpotifySelecionado.sort_values(by='Dançabilidade', ascending=False).head(10)
```

**LOAD**

```python
top10MusicasDancantes.to_csv('top10MusicasDançantes2023.csv')
```
